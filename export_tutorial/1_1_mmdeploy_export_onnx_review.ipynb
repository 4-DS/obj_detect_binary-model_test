```python
import json

config_fn = os.path.join(cache_urls.cache_data, 'config.json')

with open(config_fn) as f_id:
    CONFIG = json.load(f_id)

MODEL_NAME = CONFIG["train_config_parameters"]["MODEL_NAME"]
WORK_DIR = CONFIG['work_dir'] 
ONNX_DIR = CONFIG["onnx_dir"]
CONFIG_DEPLOY = CONFIG["config_deploy"]
DEVICE = CONFIG["device"]
CONFIG_MODEL = CONFIG["config_model"]
TORCH_MODEL = CONFIG["torch_model"]
ONNX_MODEL = CONFIG["onnx_model"]
```

```python
# TEST IMAGE
test_image_path = osp.join(cache_urls.cache_data, "test.png")
assert osp.exists(test_image_path)
```

```python
from mmdet.datasets import PIPELINES

@PIPELINES.register_module()
class DataAsList:
    def __call__(self, results):
        aug_data_dict = {key: [val] for key, val in results.items()}
        return aug_data_dict
```

## Конвертирование модели YOLOX через mmdeploy    
### препроцессинг и постпроцессинг конвертируется в ONNX модель совместно с графом модели

```python
from mmdeploy.apis import torch2onnx
import numpy as np

torch2onnx(img=np.zeros((640,640,3)),
           work_dir=ONNX_DIR,
           save_file=ONNX_MODEL,
           deploy_cfg=CONFIG_DEPLOY,
           model_cfg=CONFIG_MODEL,
           model_checkpoint=TORCH_MODEL,
           device=DEVICE)
```

### Инференс обученной модели на тестовом изображении через mmdetection

```python
from mmdet.apis import init_detector, inference_detector, show_result_pyplot
```

Инициализация модели

```python
# build the model from a config file and a checkpoint file
model = init_detector(CONFIG_MODEL, 
                      TORCH_MODEL, 
                      device='cpu')
```

Запуск инференса детектора через api mmdetection

```python
out = inference_detector(model, test_image_path)
```

Отображение предсказанного результата на изображении

```python
show_result_pyplot(model, test_image_path, out, score_thr=0.05)
```

### Инференс сконвертированной модели на тестовом изображении через ONNXRuntime

Инициализация сессии onnxruntime для сконвертированной модели

```python
import onnxruntime

output_file = os.path.join(ONNX_DIR, ONNX_MODEL)
output_file

model_onnx = onnxruntime.InferenceSession(output_file, providers=['CUDAExecutionProvider'])
input_name = model_onnx.get_inputs()[0].name
output_name = [out.name for out in model_onnx.get_outputs()]
output_name
```

Подготовка тестового изображения для onnx модели.

```python
from mmdet.datasets import replace_ImageToTensor
from mmdet.datasets.pipelines import Compose
test_pipeline = Compose(model.cfg.data.test.pipeline)

data = test_pipeline(dict(img_info=dict(filename=test_image_path), img_prefix=None))
input_data = data["img"][0].data.numpy()
input_data = input_data[np.newaxis, ...].astype(np.float32)
```

Инференс тестового изображения

```python
dets, labels = model_onnx.run(output_name, {input_name: input_data})
```

```python
show_result_pyplot(model, test_image_path, dets, score_thr=0.05)
```
